# Dichtefunktion

```python
x = np.arange(1,100.)/50.
def weib(x,n,k):
     return  (k / n) * (x / n)**(k - 1) * np.exp(-(x / n)**k)
count, bins, ignored = plt.hist(np.random.weibull(5.,1000))
scale = count.max()/weib(x, 1., 5.).max()
plt.plot(x, weib(x, 1., 1.)*scale)
plt.plot(x, weib(x, 1., 2.)*scale)
plt.plot(x, weib(x, 1., 3.)*scale)
plt.plot(x, weib(x, 1., 3.602)*scale)
plt.show()   
```

# Zuverlässigkeit

```python
x = np.arange(1,100.)/50.
def weib(x,n,k):
     return 1 - ((k / n) * (x / n)**(k - 1) * np.exp(-(x / n)**k))
count, bins, ignored = plt.hist(np.random.weibull(5.,1000))
scale = count.max()/weib(x, 1., 5.).max()
plt.plot(x, weib(x, 1., 1.)*scale)
plt.plot(x, weib(x, 1., 2.)*scale)
plt.plot(x, weib(x, 1., 3.)*scale)
plt.plot(x, weib(x, 1., 3.602)*scale)
plt.show()    
```
